```python
import nltk
import re
import os
from nltk.probability import FreqDist
from nltk.corpus import stopwords

list=[]
textt=''
#Read file text
src='D:/Tài liệu học/kì 5 năm 3/Khai phá dữ liệu web/kpw/4_12_Chapter4Final/4_12_Chapter4Final/data_preprocessing'
path=os.listdir(src)

for i in path:
    label=i
    f=open('D:/Tài liệu học/kì 5 năm 3/Khai phá dữ liệu web/kpw/4_12_Chapter4Final/4_12_Chapter4Final/data_preprocessing/'+str(label), "r", encoding="utf-8")
    text=f.read()
    textt=textt+text
# print(textt)
text_pre=[]
text_pre=textt.split(" ")

frequency_dist = FreqDist(word.lower() for word in text_pre)
    ## show only th top 50 results
print(frequency_dist.most_common(50))
    ## Consider words with length greater than 3 and plot
large_words = dict([(k,v) for k,v in frequency_dist.items() if len(k)>3])
frequency_dist = nltk.FreqDist(large_words)
frequency_dist.plot(50,cumulative=False)


from wordcloud import WordCloud
wcloud = WordCloud().generate_from_frequencies(frequency_dist)
    #plotting the wordcloud
import matplotlib.pyplot as plt
plt.imshow(wcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
    
        
```
